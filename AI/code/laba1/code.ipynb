#### Мінімаксна інтерпретація логічних операторів з використанням операцій пошуку мінімуму і максимуму

```python
def min_f(x, f1, f2):
    x_min, y_min = skfuzzy.fuzzy_and(x, f1, x, f2)
    plt.plot(x, f1, linestyle='--')
    plt.plot(x, f2, linestyle='--')
    plt.plot(x_min, y_min)
    plt.title(f"Min")
    plt.show()


x = np.arange(-20, 25, 0.5)
min_f(x, skfuzzy.gaussmf(x, 0, 5), skfuzzy.gaussmf(x, 5, 5))
```

```python
def max_f(x, f1, f2):
    x_max, y_max = skfuzzy.fuzzy_or(x, f1, x, f2)
    plt.plot(x, f1, linestyle='--')
    plt.plot(x, f2, linestyle='--')
    plt.plot(x_max, y_max)
    plt.title('Max')
    plt.show()


x = np.arange(-20, 25, 0.5)
max_f(x, skfuzzy.gaussmf(x, 0, 5), skfuzzy.gaussmf(x, 5, 5))
```

## Вірогідна інтерпритація кон’юктивну і диз’юктивних операторів

```python
def prob_min_f(x, f1, f2):
    y = f1 * f2
    plt.plot(x, y)
    plt.plot(x, f1, linestyle='--')
    plt.plot(x, f2, linestyle='--')
    plt.title(f"Probable Min")
    plt.show()


x = np.arange(-20, 25)
prob_min_f(x, skfuzzy.gaussmf(x, 0, 5), skfuzzy.gaussmf(x, 5, 5))
```

```python
def prob_max_f(x, f1, f2):
    y = f1 + f2 - (f1 * f2)
    plt.plot(x, y)
    plt.plot(x, f1, linestyle='--')
    plt.plot(x, f2, linestyle='--')
    plt.title(f"Probable Max")
    plt.show()


x = np.arange(-20, 20)
prob_max_f(x, skfuzzy.gaussmf(x, 0, 5), skfuzzy.gaussmf(x, 5, 5))
```

# Заперечення

```python
def not_f(x, func):
    not_func = skfuzzy.fuzzy_not(func)
    plt.plot(x, func, linestyle='--')
    plt.plot(x, not_func)
    plt.title("NOT")
    plt.show()


x = np.arange(-20, 20)
not_f(x, skfuzzy.gaussmf(x, 0, 8))
```
