## NetworkX

Here i am going to use the networkx library for python, and build some graph systems

```python
import networkx as nx

g = nx.DiGraph()
```

#### Adding nodes

```python
g.add_node('Alice')
g.add_node('Bob')
g.add_node('John')

g.nodes()
```

#### Adding edges

```python
g.add_edge('Alice', 'Bob')
g.add_edge('John', 'Bob')

g.edges()
```

#### Drawing 
Here we can see how to draw graph, this function will draw it in circular way

```python
nx.draw_circular(g, node_size=2000, with_labels=True)
```

Lets write cities representation

```python
from math import sqrt
from itertools import combinations

def find_dist(coor1, coor2):
    return round(sqrt((coor1[0] - coor2[0])**2 + (coor1[1] - coor2[1])**2), 1)
    
coordinates = {
    'Kiev': (-49, 20),
    'Chernigov': (-20, 100),
    'Vinnica': (-60, -100)
}
city_combinations = combinations(coordinates.keys(), 2)


cities_comb_with_dist = {
    edge: find_dist(coordinates[edge[0]], coordinates[edge[1]]) for edge in city_combinations
}

cities = nx.Graph()
cities.add_edges_from(cities_comb_with_dist)

nx.draw(cities, coordinates,
       node_size=200,
       with_labels=True,
       node_color=(10, 1, 10))

nx.draw_networkx_edge_labels(cities, coordinates,
                             cities_comb_with_dist)
print()
```
